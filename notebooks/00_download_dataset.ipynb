```python
import zipfile
from pathlib import Path

import gdown
```

```python
RAW = Path("../data/raw")
RAW.mkdir(parents=True, exist_ok=True)
```

```python
URLS = {
    "Training.zip": "https://drive.google.com/file/d/1d4HfISGKdKyCvThw7B9rvSaBWRrCYyog/view?usp=sharing",
    "Test.zip": "https://drive.google.com/file/d/1FZJemwjfAiv95LYLLYdBiInIfdi-47Rm/view?usp=sharing",
    "Masks.zip": "https://drive.google.com/file/d/1C5R7h8s0qF-7MQS8Oz3hkLHqM7vzITKm/view?usp=sharing",
}
```

```python
for name, url in URLS.items():
    dest = RAW / name
    if dest.exists():
        print(f"[SKIP] {name} already exists")
        continue
    gdown.download(url, str(dest), fuzzy=True)
```

```python
for name in URLS:
    zippath = RAW / name
    extract_to = RAW / name.replace(".zip", "")
    if extract_to.exists():
        print(f"[SKIP] {extract_to.name} already extracted")
        continue
    with zipfile.ZipFile(zippath, "r") as zf:
        zf.extractall(RAW)
    print(f"[OK] Extracted {name}")
```
